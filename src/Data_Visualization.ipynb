```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
# Jitter plot comparing the correlation between Credit Score and the Outcome (car insurance claimed or not)
# The results show that higher credit score decrease car insurance claims, but very minimally
# The graph directly aligns with the logistic regression coefficient for credit score
sns.stripplot(x=df['OUTCOME'], y=df['CREDIT_SCORE'], data=df, jitter=True, dodge=True, alpha=0.1)
plt.title('Credit Score vs Outcome')
plt.xlabel('OUTCOME')
plt.ylabel('CREDITSCORE')
plt.savefig('credit_score_vs_outcome.png')
plt.show()
```

```python
# Bar plot comparing driving experience with speeding violations
# Results show that drivers with more years of driving experience generally have more speeding violations
# A mediating variable may be time, not necessarily driving experience
sns.barplot(x=df['DRIVING_EXPERIENCE'], y=df['SPEEDING_VIOLATIONS'])
plt.title('Driving Experience vs Speeding Violations')
plt.xlabel('DRIVING_EXPERIENCE')
plt.ylabel('SPEEDING_VIOLATIONS')
plt.savefig('driving_experience_vs_speeding_violations.png')
plt.show()
```

```python
# The bar plot shows that fewer years of driving experience generally results
# in a higher likelihood of filing car insurance claims.
sns.barplot(x=df['DRIVING_EXPERIENCE'], y=df['OUTCOME'])
plt.title('Driving Experience vs Outcome')
plt.xlabel('DRIVING_EXPERIENCE')
plt.ylabel('OUTCOME')
plt.savefig('driving_experience_vs_outcome.png')
plt.show()
```

```python
# Line plot shows that the number of speeding violations have a possitive
# correlation with car insurance claims. The high percentage of claims for
# drivers with 0 speeding violations may be due to the large majority of
# drivers not having speeding violations.
sns.lineplot(x=df['SPEEDING_VIOLATIONS'], y=df['OUTCOME'])
plt.title('Speeding Violations vs Outcome')
plt.xlabel('SPEEDING_VIOLATIONS')
plt.ylabel('OUTCOME')
plt.savefig('speeding_violations_vs_outcome.png')
plt.show()
```

```python
# The following line plot shows how people with more DUIs generally have higher
# odds of claiming car insurance, though it peaks at around the same level of
# 0 DUIs. This is due to the majority of drivers having 0 DUIs.
sns.lineplot(x=df['DUIS'], y=df['OUTCOME'])
plt.title('DUIS vs Outcome')
plt.xlabel('DUIS')
plt.ylabel('OUTCOME')
plt.savefig('duis_vs_outcome.png')
plt.show()

# As shown in this table, 6,606 of the drivers out of the
df['DUIS'].value_counts()
```

```python
# The barplot below shows how drivers who own vehicles made prior to 2015 have
# a much higher chance of getting into accidents, thus a higher chance of
# filing an auto insurance claim.
sns.barplot(x=df['VEHICLE_YEAR'], y=df['OUTCOME'])
plt.title('Vehicle Year vs Outcome')
plt.xlabel('VEHICLE_YEAR')
plt.savefig('vehicle_year_vs_outcome.png')
plt.ylabel('OUTCOME')
plt.show()
```
